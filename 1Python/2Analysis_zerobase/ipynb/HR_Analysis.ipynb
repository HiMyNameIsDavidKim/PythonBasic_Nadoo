# 사내 조직 파악하기

## Group By

```python
tmp0 = df[['Department', 'EducationField', 'JobRole', 'JobLevel', 'Attrition']].copy()
```

```python
tmp0.groupby(['Department']).size()
```

```python
tmp0.groupby(['Department', 'JobRole']).size().sort_values(ascending=False)
```

```python
tmp1 = df[['Department', 'JobRole', 'JobLevel', 'Age', 'Attrition']].copy()
```

```python
tmp1.groupby(['Department', 'JobRole'])['Age'].agg(['mean', 'max', 'min', 'count'])
```

## Pivot Table

```python
pt1 = pd.pivot_table(tmp1,
               index=['Attrition', 'Department'],
               columns='JobLevel',
               values='Age',
               aggfunc='count',
               fill_value='',
               )

pt1
```

## Unstack, Stack

```python
pt2 = pd.pivot_table(tmp1,
               index=['Attrition', 'Department', 'JobRole'],
               columns='JobLevel',
               values='Age',
               aggfunc='count',
               fill_value='',
               )

pt2
```

```python
# 0: 'Attrition', 1: 'Department', 2: 'JobRole'

st1 = pt2.unstack(level=0).stack(level=0)
st1
```

# Visualization

```python
df.corr(numeric_only=True)
```

```python
cor_df = df.select_dtypes(include='int64').corr(numeric_only=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(cor_df,
            annot=True,
            fmt='.1f',
            linewidths=0.5,
            cmap='YlGnBu',
            )
```

```python
top8_cols = df.select_dtypes(include='int64').corr().sum().sort_values(ascending=False)[:8].index.tolist()
top8_cols
```

```python
cor_df2 = df[top8_cols].corr()
sns.heatmap(cor_df2,
            annot=True,
            fmt='.1f',
            linewidths=0.5,
            cmap='Blues',
)
```

```python
mask = np.zeros_like(cor_df2, dtype=bool)
mask[np.triu_indices_from(mask)] = True
mask
```

```python
sns.heatmap(cor_df2,
            annot=True,
            fmt='.1f',
            linewidths=0.5,
            cmap='Blues',
            mask=mask,
)
```

```python
# Simple version - Matplotlib
plt.title('Simple Bar plot - Income by JobLevel')
plt.bar(df['JobLevel'], df['MonthlyIncome']) #plt.bar(x, y)
plt.xlabel('JobLevel')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Simple version - Matplotlib
plt.title('Simple Bar plot - Income by TotalWorkingYears')
plt.bar(df['TotalWorkingYears'], df['MonthlyIncome']) #plt.bar(x, y)
plt.xlabel('TotalWorkingYears')
plt.ylabel('Monthly Income')
plt.show()
```

```python
facet = sns.FacetGrid(df, col='JobLevel', row='Gender', height=3)
facet.map_dataframe(sns.barplot, x='YearsAtCompany', y='MonthlyIncome')
plt.show()
```

```python
# Attrition Yes와 No에 어떤 차이가 있나? -- hue 추가
facet = sns.FacetGrid(df, col='Department', row='Gender', hue='Attrition', height=5)
facet = facet.map_dataframe(sns.regplot, x='Age', y='MonthlyIncome')
facet = facet.add_legend()
```

```python
facet = sns.FacetGrid(df, col='Department', row='Gender', hue='Attrition', height=5, hue_order=['No','Yes'], palette={'Yes':'red', 'No':'gray'})
facet = facet.map_dataframe(sns.regplot, x='Age', y='MonthlyIncome', fit_reg=False)
facet = facet.add_legend()
```
